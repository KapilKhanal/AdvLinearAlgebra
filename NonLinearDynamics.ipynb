```python
# import necessary libraries
import matplotlib.pyplot as plt
# show plots in notebook
% matplotlib inline

# define system in terms of separated differential equations
def f(S,I,alpha):
    return -alpha * S*I
def g(S,I,alpha,b ):
    return alpha*S*I - b*I

# initialize lists containing initial values
x = [800,]
y = [800,]

# dt = timestep, N = range
def sys( dt, N,alpha):
    # compute and fill lists
    for i in range(N):
        x.append(x[i] + (f(x[i],y[i],2.2*10**(-3))) * dt)
        y.append(y[i] + (g(x[i],y[i],2.2*10**(-3),0.4)) * dt)
        #z.append(z[i] + (h(x[i],y[i],z[i])) * dt)
    return x, y

sys(0.01, 800,2.2)

#plot
fig = plt.figure(figsize=(15,5))
fig.subplots_adjust(wspace = 0.5, hspace = 0.3)
ax2 = fig.add_subplot(1,2,1)

ax2.plot(x, y, color="blue")
ax2.set_xlabel("x")
ax2.set_ylabel("y")  
ax2.set_title("Phase space")
ax2.grid()
```
